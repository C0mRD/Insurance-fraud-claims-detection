```python
import pandas as pd
```

```python
df = pd.read_csv("../data/after_ingestion_insurance.csv")
```

Check how many unique values each column has ??

```python
def check_unique_values(df):
    col_names = []
    col_unique_values = []
    for col in df:
        if df[col].dtype == 'object':
            col_names.append(str(col))
            col_unique_values.append(df[col].nunique())
    unique_val_table = pd.DataFrame()
    unique_val_table["Col Name"] = col_names
    unique_val_table["Unique Values"] = col_unique_values
    unique_val_table = unique_val_table.sort_values(by='Unique Values', ascending=False)
    return unique_val_table
```

```python
unique_table = check_unique_values(df)
unique_table
```
